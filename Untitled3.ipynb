```python
IMG_SIZE = 128

def load_data(image_list, mask_list):

    X=[]
    Y=[]

    for img_path,mask_path in zip(image_list,mask_list):

        img=cv2.imread(img_path)
        img=cv2.resize(img,(IMG_SIZE,IMG_SIZE))
        img=img/255.0

        mask=cv2.imread(mask_path,0)
        mask=cv2.resize(mask,(IMG_SIZE,IMG_SIZE))
        mask=np.expand_dims(mask,-1)
        mask=mask/255.0

        X.append(img)
        Y.append(mask)

    return np.array(X),np.array(Y)
```

```python
import time

print("Loading training data...")
start_time = time.time()
X_train,Y_train = load_data(train_images,train_masks)
end_time = time.time()
print(f"Training data loaded in {end_time - start_time:.2f} seconds.")

print("Loading validation data...")
start_time = time.time()
X_val,Y_val     = load_data(val_images,val_masks)
end_time = time.time()
print(f"Validation data loaded in {end_time - start_time:.2f} seconds.")

print("Loading test data...")
start_time = time.time()
X_test,Y_test   = load_data(test_images,test_masks)
end_time = time.time()
print(f"Test data loaded in {end_time - start_time:.2f} seconds.")

print("\nTrain shape:",X_train.shape)
print("Validation shape:",X_val.shape)
print("Test shape:",X_test.shape)
```

```python
def conv_block(x,filters):

    x=Conv2D(filters,3,padding="same")(x)
    x=BatchNormalization()(x)
    x=Activation("relu")(x)

    x=Conv2D(filters,3,padding="same")(x)
    x=BatchNormalization()(x)
    x=Activation("relu")(x)

    return x


def build_unet():

    inputs=Input((IMG_SIZE,IMG_SIZE,3))

    c1=conv_block(inputs,64)
    p1=MaxPooling2D()(c1)

    c2=conv_block(p1,128)
    p2=MaxPooling2D()(c2)

    c3=conv_block(p2,256)
    p3=MaxPooling2D()(c3)

    c4=conv_block(p3,512)
    p4=MaxPooling2D()(c4)

    c5=conv_block(p4,1024)

    u6=UpSampling2D()(c5)
    u6=Concatenate()([u6,c4])
    c6=conv_block(u6,512)

    u7=UpSampling2D()(c6)
    u7=Concatenate()([u7,c3])
    c7=conv_block(u7,256)

    u8=UpSampling2D()(c7)
    u8=Concatenate()([u8,c2])
    c8=conv_block(u8,128)

    u9=UpSampling2D()(c8)
    u9=Concatenate()([u9,c1])
    c9=conv_block(u9,64)

    outputs=Conv2D(1,1,activation="sigmoid")(c9)

    model=Model(inputs,outputs)

    return model


model=build_unet()
model.summary()
```

```python
def dice_coef(y_true,y_pred):

    y_true=tf.reshape(y_true,[-1])
    y_pred=tf.reshape(y_pred,[-1])

    intersection=tf.reduce_sum(y_true*y_pred)

    return (2*intersection+1)/(tf.reduce_sum(y_true)+tf.reduce_sum(y_pred)+1)


def dice_loss(y_true,y_pred):

    return 1-dice_coef(y_true,y_pred)
```

```python
model.compile(

optimizer=Adam(1e-4),
loss=dice_loss,
metrics=[dice_coef]

)

print("Model Compiled Successfully")
```

```python
history = model.fit(

X_train,Y_train,
validation_data=(X_val,Y_val),
epochs=25,
batch_size=8

)
```
